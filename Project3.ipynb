```python
import pandas as pd
import json
```

```python
dataList = []
with open(r"C:\Users\donni\Documents\work\2023_1\BDA\supplies.json") as f:
    for jsonObj in f:
        dataDict = json.loads(jsonObj)
        dataList.append(dataDict)

df = pd.json_normalize(dataList)
```

```python
df
```

#### Tranformation

Flatten "items" column. Sales ID is added so it may be used for merging with the main dataframe.

```python
df_items = pd.json_normalize(dataList,record_path="items", meta=[['_id','$oid']])

df_items
```

Drop "items" column for the dataframe, then merge with the dataframe from previous step.

```python
df2 = df.drop(columns="items")

df3 = pd.merge(df2, df_items, on="_id.$oid")
df3
```

Cleaning up the dataframe.

```python
df3.rename(columns= {"_id.$oid":"id", \
                     "saleDate.$date":"date",
                     "customer.gender":"gender",
                     "customer.age":"age",
                     "customer.email":"email",
                     "customer.satisfaction":"satisfaction",
                     "name":"item",
                     "price.$numberDecimal":"price"}, inplace=True)

df3["date"] = df3["date"].str.slice(start=0,stop=10)
df3["price"] = df3["price"].astype("float")
df3
```

```python
df3.info()
```

### Query

Show top 10 products (name) sales (quantity x price).

```python
df4 = df3.assign(sales=lambda x: x.quantity * x.price)
df_10sales = df4.groupby('item').sum()['sales'].reset_index()
df_10sales = df_10sales.sort_values('sales',ascending = False)
df_10sales[:10]
```

Show top 3 products (name) sales by store (location).

```python
df_storetop3 = df4.groupby(["storeLocation","item"]).sum("sales").reset_index()\
    .drop("age",axis="columns")\
    .drop("couponUsed",axis="columns")\
    .drop("quantity",axis="columns")\
    .drop("price",axis="columns")\
    .drop("satisfaction",axis="columns")
```

```python
df_storetop3.set_index(['storeLocation','item']).sales.groupby('storeLocation',group_keys=False).nlargest(3)
```

Show rankings of each store (location).

```python
df_ranking = df.groupby('storeLocation').sum()['customer.satisfaction'].reset_index()
df_ranking = df_ranking.sort_values('customer.satisfaction',ascending = False)
df_ranking[:10]
```

Show purchased method by gender table 

```python
df_method = df[["customer.gender","purchaseMethod"]].groupby(["customer.gender"]).value_counts()
df_method.head()
```

Show monthly total sales

```python
df4["date"] = pd.to_datetime(df4["date"], dayfirst=False)
df_monthlysales = df4[["date","sales"]].groupby([pd.Grouper(freq='M',key="date")]).sum('sales')
df_monthlysales
```
